```python
# import dependencies 
import pandas as pd
import numpy as np


import warnings
warnings.filterwarnings("ignore")
```

```python
#reading data
song_numerics = pd.read_csv('data/song_data.csv') 
song_genre = pd.read_csv('data/song_info.csv') 
```

```python
#Convert all string  to lower case
song_genre['playlist'] = song_genre['playlist'].str.lower()
song_genre['song_name'] = song_genre['song_name'].str.lower()
song_genre['artist_name'] = song_genre['artist_name'].str.lower()
song_genre['album_names'] = song_genre['album_names'].str.lower()
song_numerics['song_name'] = song_numerics['song_name'].str.lower()
```

```python
song_numerics = song_numerics.drop('song_duration_ms', axis=1)
song_numerics = song_numerics.drop('time_signature', axis=1)

```

```python
# Merging the two datasets into one
song_numerics_with_artists = song_numerics.copy()

#add artist name so that you can drop the songs that are multiple playlists
song_numerics_with_artists['artist_name'] = song_genre['artist_name']

#droping duplicates (there are duplicates bc the same song is are in multiple different playlists in song info)
song_numerics_with_artists = song_numerics_with_artists.drop_duplicates(subset=['song_name', 'artist_name'])

# reset index 
song_numerics_with_artists = song_numerics_with_artists.reset_index()
song_numerics_with_artists = song_numerics_with_artists.drop('index',axis=1)

song_numerics_with_artists.head()
```

```python
# Standardise using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

minmaxscaler = MinMaxScaler()

# leave out artist normalization
minmaxscaled = minmaxscaler.fit_transform(song_numerics_with_artists[['song_popularity',
                                                                       #'song_duration_ms', 
                                                                       'acousticness',
                                                                       'danceability',
                                                                       'energy',
                                                                       'instrumentalness',
                                                                       'key',
                                                                       'liveness',
                                                                       'loudness',
                                                                       'audio_mode',
                                                                       'speechiness',
                                                                       'tempo',
                                                                       #'time_signature',
                                                                       'audio_valence']])

songs_normalized = pd.DataFrame(minmaxscaled, columns=[['song_popularity', 
                                                        #'song_duration_ms',
                                                        'acousticness',
                                                        'danceability',
                                                        'energy',
                                                        'instrumentalness',
                                                        'key',
                                                        'liveness',
                                                        'loudness',
                                                        'audio_mode', 
                                                        'speechiness', 
                                                        'tempo', 
                                                        #'time_signature',
                                                        'audio_valence']])
```

```python
# create feature matrix
song_features=song_df_normalized.set_index("song_name")
song_features.drop(["artist_name"], axis=1, inplace=True)
song_features.head()
```

```python
#save both df into csv files 

song_features.to_csv('song_features.csv')
song_genre_nodup=song_genre_nodup.set_index("song_name")
song_genre_nodup.to_csv('song_genre.csv')

```

```python
# Use cosine similarity based NN algorithm to calculate min distances and find similar songs
from scipy.sparse import csr_matrix

song_features_csr = csr_matrix(song_features.values)
```

```python
from sklearn.neighbors import NearestNeighbors

model_nn = NearestNeighbors(metric='cosine',algorithm='brute')
model_nn.fit(song_features_csr)
```

```python
song_features=pd.read_csv("song_features.csv")
temp_genre=pd.read_csv("song_genre.csv")

```

```python
temp_genre.head()
```

```python
temp = song_features.copy()
temp.reset_index(inplace=True)
songsearch = 'let it be'
songsearch = songsearch.lower()
song_index = temp.index[temp['song_name'] ==songsearch].tolist()[0]




distances,indices = model_nn.kneighbors(X = temp.iloc[song_index,2:].values.reshape(1,-1), n_neighbors=6)


for i in range(0, len(distances.flatten())):
    if i == 0:
        print("Recommendation for", temp.iloc[song_index, 1], "by", temp_genre.iloc[song_index, 1], "are: ")
    else:
        print(i,": ",temp_genre.iloc[indices.flatten()[i], 0], "by", temp_genre.iloc[indices.flatten()[i], 1], "|  distance= ",distances.flatten()[i])

```
